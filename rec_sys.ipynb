```python
# from here i deicded to create a pivot table using ratings.
# after pivot table ill depict coorelation about the movies, 
# considering a dummy user that watches A title of movie/so now what are the chances that the 
# user may get recommended from his previous recoreded ratings
#  using corrwith (correlation)
```

```python
ratings['num of ratings'] = pd.DataFrame(dataset.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=20)
```

```python

plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
moviemat = dataset.pivot_table(index='userId',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
movie1 = moviemat['Terminator 2: Judgment Day (1991)']
# movie2 = moviemat['Terminator 2: Judgment Day (1991)']
movie1.head()
```

```python
similar_movie1 = moviemat.corrwith(movie1)
# similar_movie2 = moviemat.corrwith(movie2)
```

```python
corr_movie1 = pd.DataFrame(similar_movie1, columns=['Correlation'])
corr_movie1.dropna(inplace=True)
corr_movie1.head()
```

```python
corr_movie1.sort_values('Correlation', ascending=False).head()
```

```python
corr_movie1 = corr_movie1.join(ratings['num of ratings'])
corr_movie1.head()
```

```python
result = corr_movie1[corr_movie1['num of ratings']>100].sort_values('Correlation',ascending=False)
result.head()
```

```python
print("So top recommended movies are: ")
for i in range(10):
    print("{} : {}".format(i+1, result.index[i]))
```
